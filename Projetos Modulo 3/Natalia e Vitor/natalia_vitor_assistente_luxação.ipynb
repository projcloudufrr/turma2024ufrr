```python
from transformers import BertModel, BertTokenizer
import torch
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def gerar_embeddings(setencas):
  # Carregar o modelo BERT pré-treinado e o tokenizador
  modelo = BertModel.from_pretrained('bert-base-multilingual-cased', output_hidden_states=True)
  tokenizador = BertTokenizer.from_pretrained('bert-base-multilingual-cased')
  # Calcular os embeddings BERT para cada sentença
  embeddings = []
  for sentenca in setencas:
    tokens = tokenizador.tokenize(sentenca)
    ids = tokenizador.convert_tokens_to_ids(tokens)
    tensor = torch.tensor([ids])
    with torch.no_grad():
        saida = modelo(tensor)
    vetor_embedding = saida[2][-1].mean(dim=1).squeeze().numpy()
    embeddings.append(vetor_embedding)
  return embeddings
```

```python
def diagnosticar(sintoma_paciente):
    # Tokenize the symptoms
    tokenizador = BertTokenizer.from_pretrained('bert-base-multilingual-cased')
    tokens = tokenizador.tokenize(sintoma_paciente)
    ids = tokenizador.convert_tokens_to_ids(tokens)

    # Generate embeddings
    embedding_paciente = gerar_embeddings([ids])
    embedding_paciente = np.array(embedding_paciente).reshape(1, -1)

    # Calculate cosine similarities
    similaridades = []
    for embedding in df['Embeddings']:
        embedding = np.array(embedding).reshape(1, -1)
        similaridade = cosine_similarity(embedding_paciente, embedding)[0][0]
        similaridades.append(similaridade)

    # Find the most similar diagnosis
    similaridades = pd.Series(similaridades)
    idx_max_similaridade = similaridades.idxmax()

    # Return the result
    resultado = df.loc[idx_max_similaridade, ["Possiveis Vértebra com subluxações", "Área Corporal"]]
    resultado = resultado.to_frame().reset_index()
    resultado.columns = ['Descrição', 'Detalhe']
    return resultado
```

```python
tabs = gr.Interface(
    fn=diagnosticar,
    inputs = "text",
    outputs = "dataframe",
    title = "Assistente Diagnostico de Subluxações Para Quiropraxista",
    description = "Digite os sintomas do paciente:"
)

tabs.launch(debug=True)
```
